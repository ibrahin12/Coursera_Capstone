```python
# importing necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
# getting data from internet
buffalo_link='https://data.mongabay.com/igapo/zip_codes/metropolitan-areas/metro-zip/Buffalo-Niagara%20Falls%20(NY)1.html'
raw_buffalo_page= requests.get(buffalo_link).text

# using beautiful soup to parse the HTML/XML codes.
soup = BeautifulSoup(raw_buffalo_page)
#print(soup.prettify())
```

```python
# extracting the raw table inside that webpage
table = soup.find('table', {'class':'boldtable'}) 

df_list= []
# extracting a clean form of the table
for tr_cell in soup.find_all('tr'):
    for td_cell in tr_cell.find_all('td'):
        df_list.append(td_cell)

df_list
```
