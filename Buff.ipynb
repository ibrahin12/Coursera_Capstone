```
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
pd.set_option("display.max_rows",999)
```

```
# getting data from internet
buffalo_link='https://data.mongabay.com/igapo/zip_codes/metropolitan-areas/metro-zip/Buffalo-Niagara%20Falls%20(NY)1.html'
raw_buffalo_page= requests.get(buffalo_link).text

# using beautiful soup to parse the HTML/XML codes.
soup = BeautifulSoup(raw_buffalo_page)
#print(soup.prettify())
```

```
table = soup.find('table', {'class':'boldtable'}) 

df_list= []
# extracting a clean form of the table
for tr_cell in soup.find_all('tr'):
    for td_cell in tr_cell.find_all('td'):
        df_list.append(td_cell.text.strip())

df_list[123:-2]

df_list=df_list[123:-2]
```

```
zipTown = df_list[0::6]
zipTown = zipTown[232:]
areaCode = df_list[1::6]
areaCode = areaCode[232:]
county = df_list[2::6]
county = county[232:]
state = df_list[3::6]
state = state[232:]
metroArea = df_list[4::6]
metroArea = metroArea[232:]
etcList = df_list[5::6]
etcList = etcList[232:]
```

```
zipCodes = []
town = []
for i in zipTown:
    zips = i[0:5]
    towns = i[6:]
    zipCodes.append(zips)
    town.append(towns)
```

```
metroArea2 = []
for j in metroArea:
    metar = j[0:25]
    metroArea2.append(metar)
```

```
df_buffalo = pd.DataFrame()
```

```
df_buffalo['Zip Codes'] = zipCodes
df_buffalo['Town'] = town
df_buffalo['Area Code'] = areaCode
df_buffalo['County'] = county
df_buffalo['State'] = state
df_buffalo['Metropolitan Area'] = metroArea2
```

```
df_buffalo=df_buffalo[:197]
```
